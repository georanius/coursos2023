# <b> Data analysis

In the following some basic analysis is applied to the data set for demonstration purposes.

Important:<br>
Any (more) advanced analysis might require a prior check if the data set fulfills the requirements of the analysis, e.g. the data need to be normal distributed or must contain a minimum number of independent data points and so on. The technical feasibility alone does not allow the conclusion that an analysis technique is appropriate or sensible (remember we calculated the surface of a negative radius...).

````python
# renaming the columns for convenience: informative varible_names
# or wasting memory space... if col_data 

years = col_data[:,0]
carb = col_data[:,1]
````

__Objective:__ e.g. to evaluate if the level is much higher today than in these 800 kyr before


__Task:__ find the maximal level of carbon dioxid in the last 800 kyr

````python
max_carb = np.max(carb)
print(max_carb)

max_year = years[np.where(carb == max_carb)]
print(max_year)
````

```np.where``` is a useful function to mask an array in order to find the indices of values that fulfil a given boolean condition.

__Objective:__ what's the typical annual variation?


__Task:__ find difference in carbon dioxide between data points

````python
difference_per_year = np.diff(carb)

plt.plot(years[:-1],difference_per_year)
````

__Objective:__ provide an estimate of mean/median/quantiles of carbon dioxide levels (i.e. obtain statistical parameters which might be the basis for further analysis, e.g. check if the data is normally distributed)


__Task:__ look at distribution of values

````python
mc = np.mean(carb)

plt.hist(carb, bins = 50)
plt.vlines(x = mc, ymin = 0, ymax = 50, colors = 'purple', label = 'mean CO2') 
plt.legend()
````

````python
# pick every k-th point to reduce the data set
k=10
plt.plot(years[::k],carb[::k],'-+')
````
